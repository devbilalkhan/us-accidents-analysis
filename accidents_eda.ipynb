```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import numpy as np
```

```python
df = pd.read_csv("data/US_Accidents_March23.csv", nrows=1000)
```

```python
df_loc = df.loc[(~df.Start_Lat.isna()) & (~df.Start_Lng.isna())]
```

```python
def create_map(df_loc, latitude, longitude, zoom):
    location_data = np.array(df_loc[["Start_Lat", "Start_Lng"]])
    marker_cluster = MarkerCluster(locations = location_data)
    world_coords = [latitude, longitude]
    world_map = folium.Map(location = world_coords, zoom_start=zoom, tiles='cartodbpositron')
    marker_cluster.add_to(world_map)
    return world_map
```

```python
map = create_map(df_loc, 39.50, -98.35, 4)
map
```
